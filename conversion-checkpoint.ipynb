```python
import numpy as np
from scipy import io
 
mat = io.loadmat('Pavia.mat')

print(mat.keys())

print(mat.values())
 
print(mat['pavia'].shape)

mat_t = np.transpose(mat['pavia'])

np.save('pavia.npy', mat_t)
```
